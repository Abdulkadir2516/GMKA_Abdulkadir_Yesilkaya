### Veriyi özetleme

```python
import yfinance as yf
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd


import warnings
warnings.filterwarnings("ignore")
```

1.  3. Ders sonunda hazırladığınız 2021 ve 2022 yıllarına ait BIST30 ve EUR/USD kur değerlerini içeren bileşke tablodan sadece ayın son işgününe ait değerleri içeren bir pivot tablo üretiniz.

```python
indirme_listesi = ["XU030.IS","EURTRY=X"]
Data = pd.DataFrame()

for i in indirme_listesi:
  data = yf.download(i, start="2021-01-01",end="2023-04-21")
  data = data["Adj Close"].rename(i)
  Data= pd.concat([Data, data], axis=1)

Data.columns =["Kapanış (TRY)", "EUR/TRY"]
Data
```

```python
Data.index = pd.to_datetime(Data.index) # convert index to datetime object
Data['ay-yıl'] = Data.index.strftime('%Y-%m') # apply strftime() method
Data
```

```python
Data_as = Data.pivot_table(values=["Kapanış (TRY)", "EUR/TRY"], index="ay-yıl", 
                           aggfunc= "last")#{np.max,npmean})
Data_as
```

2. Ay kırılımlı olarak ürettiğiniz tabloyu grafik olarak inceleyiniz.


```python
Data_as["Degisim"] = Data_as["Kapanış (TRY)"].diff(1)
Data_as["D_yon"] = Data_as.Degisim.map(lambda x: "Düşen gün" if x<0 else "Yükselen gün")
Data_as["Degisim%"] = (Data_as["Degisim"]/Data_as["Kapanış (TRY)"].shift(1))
Data_as
```

```python
fig_1 =px.bar(Data_as,x=Data_as.index, y="Degisim%",color="D_yon",
       color_discrete_map={"Düşen gün": "red", "Yükselen gün": "green"},
       text_auto=".0%")
fig_1.show()
```

3. Ders sonunda hazırladığınız 2021 ve 2022 yıllarına ait BIST30 ve EUR/USD kur değerlerini içeren bileşke tablodan sadece ayın son işgününe ait değerleri içeren bir tabloyu bir de resample yöntemi ile üretiniz.

```python
Data_as2 = Data.resample("BM", convention="end").asfreq()
Data_as2.index= Data_as2["ay-yıl"]
Data_as2 = Data_as2[["Kapanış (TRY)", "EUR/TRY"]]
Data_as2
```

```python
Data_as2["Degisim"] = Data_as2["Kapanış (TRY)"].diff(1)
Data_as2["D_yon"] = Data_as2.Degisim.map(lambda x: "Düşen gün" if x<0 else "Yükselen gün")
Data_as2["Degisim%"] = (Data_as2["Degisim"]/Data_as2["Kapanış (TRY)"].shift(1))
Data_as2
```

4. Resample yöntemi ile ürettiğiniz tabloyu grafik olarak inceleyiniz. 2. adımda elde ettiğiniz grafik ile karşılaştırarak yorumlayınız.

```python
fig_2 =px.bar(Data_as2,x=Data_as2.index, y="Degisim%",color="D_yon",
       color_discrete_map={"Düşen gün": "red", "Yükselen gün": "green"},
       text_auto=".0%")
fig_1.show()
fig_2.show()
```
