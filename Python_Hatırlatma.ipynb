```python
import yfinance as yf
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
B100 = yf.download("XU100.IS", start="2022-01-01", end="2022-12-13")
B100
```

```python
B100.head()
```

```python
B100.info()
```

```python
B100.describe()
```

```python
B100 = B100[B100.index > "2022-1-31"]
B100
```

```python
# satır silme
B100 = B100.iloc[2:] # ilk 2 satır silindi
B100
```

```python
B100 = B100.drop(["Low","High"], axis=1) # sütun silme
B100
```

```python
B100 = B100.loc[:,['Open','Adj Close']] # sütun silme
# OR
# B100 = B100[['Open','Adj Close']] # sütun silme

B100
```

```python
B100 = B100[['Adj Close']] # sütun silme
B100
```

```python
# İsim Değiştirme
B100.rename(columns = {"Adj Close": "Kapanış (Try)"})
B100.columns = ["Kapanış (TRY)"]
B100
```

```python
norm = B100.iloc[0,0]
B100["Normalize"] = B100["Kapanış (TRY)"]/norm
B100
```

# *BIST 100 Endeksi 2022 yılı başından sonuna kadar yaklaşık 2.65 katına çıktı*

```python
B100.Normalize.plot()
```

```python
plt.plot(B100.index,B100.Normalize, "r-")
```

```python
fig1 = px.line(B100, x=B100.index, y=B100.Normalize)
fig2 = px.bar(B100, x=B100.index, y=B100.Normalize)
fig3 = px.scatter(B100, x=B100.index, y=B100.Normalize)
fig1.show()
fig2.show()
fig3.show()
```

```python
fig_1 = px.line(B100, x=B100.index, y=B100.Normalize,
        width=1000, height=500,
        title = "BIST100 Endeksi 2022 - Normalize (1 ocak 2022)",
        hover_data = {"Normalize": ":.2f", "Kapanış (TRY)": ":.2f",},
        labels = {"Date" : "Tarih", "Normalize" : "BIST 100 Endeksi (TRY)"}
        )

fig_1.show()
```

```python
B100["Degisim"] = B100["Normalize"].diff(1)
B100
```

```python
B100["D_yon"] = B100["Degisim"]/2
B100
```

```python
B100["D_yon"] = B100["Degisim"].map(lambda x: "Düşen gün" if x<0 else "Yükselen gün")
B100
```

```python
B100["Degisim%"] = (B100["Degisim"]/B100["Normalize"].shift(1))*100
B100 
```

```python
px.bar(B100, y="Degisim%", color="D_yon", color_discrete_map={"Düşen gün": "red", "Yükselen gün": "green"})
```

```python
fig_3 = px.histogram(B100, x="Degisim%", nbins=10)
fig_3.show()
```
