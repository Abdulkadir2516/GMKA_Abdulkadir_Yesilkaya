```python
# satır silme
B100 = B100.iloc[2:] # ilk 2 satır silindi
B100
```

```python
B100 = B100.drop(["Low","High"], axis=1) # sütun silme
B100
```

```python
norm = B100.iloc[0,0]
B100["Normalize"] = B100["Kapanış (TRY)"]/norm
B100
```

```python
B100K = B100.merge(Kur, left_index= True, right_index=True)
B100K
```

```python
px.line(B100K, y=["Normalize","Normalize USD/TRY"])
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x = B100K.index , y= B100K["Normalize"], name="Normalize BIST 100"))

fig.add_trace(go.Scatter(x = B100K.index , y= B100K["Normalize USD/TRY"],name="Normalize USD/TRY"))

fig.update_layout(title="Normalize BIST 100 ve Normalize Kur ",
                  xaxis_title = "Tarih", xaxis_dtick="M1",
                  yaxis_title = "Kapanış TRY"

                  )


```

```python
indirme_listesi = ["XU100.IS","TRY=X"]
Veri_2023 = pd.DataFrame()

for i in indirme_listesi:
  data = yf.download(i, start="2022-01-01",end="2023-04-21")
  data = data["Adj Close"].rename(i)
  Veri_2023 = pd.concat([Veri_2023, data], axis=1)

Veri_2023.columns =["Kapanış (TRY)", "USD/TRY"]
Veri_2023
```

```python
# veri = Veri_2023[Veri_2023["Kapanış (TRY)"]>0]
veri = Veri_2023.dropna(axis=0)
veri
```

```python

veri["Normalize"] = veri["Kapanış (TRY)"]/ veri.iloc[0,0]
# veri["Normalize USD/TRY"] = veri["USD/TRY"]/ veri.iloc[0,1]
veri["Normalize USD/TRY"] = veri["USD/TRY"]/ veri.loc[veri.index[0],"USD/TRY"]
veri["Degisim"] = veri.Normalize.diff(1)
veri["D_yon"] = veri.Degisim.map(lambda x: "Düşen gün" if x<0 else "Yükselen gün")
veri["Degisim%"] = (veri.Degisim/veri.Normalize)*100
veri
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x = veri.index , y= veri["Normalize"], name="Normalize BIST 100"))

fig.add_trace(go.Scatter(x = veri.index , y= veri["Normalize USD/TRY"],name="Normalize USD/TRY"))

fig.update_layout(title="Normalize BIST 100 ve Normalize Kur ",
                  xaxis_title = "Tarih", xaxis_dtick="M1",
                  yaxis_title = "Kapanış TRY"

                  )


```

# **3. Bölüm**

```python
veri.index = pd.to_datetime(veri.index) # convert index to datetime object
veri['ay-yıl'] = veri.index.strftime('%Y-%m') # apply strftime() method
veri
```

```python
veri_as = veri.pivot_table(values=["Kapanış (TRY)", "USD/TRY"], index="ay-yıl", 
                           aggfunc= "last")#{np.max,npmean})
veri_as
```

```python
veri_as["Degisim"] = veri_as["Kapanış (TRY)"].diff(1)
veri_as["D_yon"] = veri_as.Degisim.map(lambda x: "Düşen gün" if x<0 else "Yükselen gün")
veri_as["Degisim%"] = (veri_as["Degisim"]/veri_as["Kapanış (TRY)"].shift(1))
veri_as
```

```python
px.bar(veri_as,x=veri_as.index, y="Degisim%",color="D_yon",
       color_discrete_map={"Düşen gün": "red", "Yükselen gün": "green"},
       text_auto=".0%")
```

```python

# alternatif bir yöntem olarak işe yaradığından resample metodu kullanalım
veri_as2 = veri.resample("BM", convention="end").asfreq()

veri_as2
```

```python
px.bar(veri_as2,x=veri_as2["ay-yıl"], y=veri_as2["Degisim%"],color="D_yon",
       color_discrete_map={"Düşen gün": "red", "Yükselen gün": "green"},
       text_auto=".0%")
```
