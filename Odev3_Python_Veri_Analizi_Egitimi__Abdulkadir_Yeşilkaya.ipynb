### 2021 EUR/TRY kur tablosunun incelenmesi

1. <i> Yahoo Finance</i> sitesinden 2021 yılına ait EUR/TRY kur değerlerini indiriniz

```python
Kur = yf.download("EURTRY=X", start="2021-01-01", end="2022-01-01")
Kur
```

2. Veriyi özet metodları ile inceleyiniz

```python
Kur.info()
```

```python
Kur.describe()
```

3. Veri temizleme: <br>
	• Tabloda sadece "Adj Close" (düzeltilmiş kapanış) sütununu bırakınız.<br>
	• "Adj Close" sütununu " Kapanış (TRY)" olarak yeniden adlandırınız.

```python
Kur = Kur[['Adj Close']] # sütun silme
Kur.columns = ["Kapanış (TRY)"]
Kur
```

4. Tabloya yeni sütun ekleme:<br>
	• Tabloya yeni bir sütun ekleyerek Kapanış (TRY) değerlerini 1 Ocak 2021 tarihindeki kapanış değerine göre normalize ediniz.<br>
	• BIST30 endeksinin 2021 yılı boyunca değişimini yorumlayınız.

```python
Kur["Normalize"] = Kur["Kapanış (TRY)"]/ Kur.loc[Kur.index[0],"Kapanış (TRY)"]
Kur
```

```python
B30 = yf.download("XU030.IS", start="2021-01-01", end="2021-12-31")
B30 = B30[["Adj Close"]]
B30.columns = ["Kapanış (TRY)"]
B30["Normalize"] = B30["Kapanış (TRY)"]/ B30.iloc[0,0]

fig_1 = px.line(B30, x=B30.index, y=B30.Normalize,
        width=1000, height=500,
        title = "BIST30 Endeksi 2021 - Normalize (1 ocak 2021)",
        hover_data = {"Normalize": ":.2f", "Kapanış (TRY)": ":.2f",},
        labels = {"Date" : "Tarih", "Normalize" : "BIST 30 Endeksi (TRY)"}
        )
fig_1.show()
```

# *BITS 30 Endeksi Yıl sonunda 1.23 kat artış olduğu görünmektedir ayrıca yıl içerisinde dengeli bir dağılım varken yıl sonuna doğru ani bir artış gözükmektedir*

5. Veriyi grafik olarak inceleyiniz:<br>
	• Tabloyu px grafiği olarak görselleştiriniz.<br>
	• Tabloyu go. grafiği olarak görselleştiriniz.

```python
fig_2 = px.line(Kur, x=Kur.index, y=Kur.Normalize,
        width=1000, height=500,
        title = "EUR/TL 2021 - Normalize (1 ocak 2021)",
        hover_data = {"Normalize": ":.2f", "Kapanış (TRY)": ":.2f",},
        labels = {"Date" : "Tarih", "Normalize" : "BIST 100 Endeksi (TRY)"}
        )

fig_2.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = Kur.index , y= Kur["Normalize"], name="Normalize EUR/TRY"))

fig.update_layout(title="Normalize EUR/TRY",
                  xaxis_title = "Tarih", xaxis_dtick="M1",
                  yaxis_title = "Kapanış TRY"

                  )
```

### 2021 ve 2022 değerlerini birlikte inceleme

1. <i> Yahoo Finance</i> sitesinden 2022 BIST30 ve EUR/TRY kur değerlerini indiriniz

```python
indirme_listesi = ["XU030.IS","EURTRY=X"]
Veri = pd.DataFrame()

for i in indirme_listesi:
  data = yf.download(i, start="2022-01-01",end="2023-01-01")
  data = data["Adj Close"].rename(i)
  Veri= pd.concat([Veri, data], axis=1)

Veri.columns =["Kapanış (TRY)", "USD/TRY"]
Veri
```

2. Indirdiğiniz verileri 2021 verilerini temizleyip düzenlediğiniz üzere hazırlayınız.

```python
# Veri = Veri[Veri.index > "2021-01-01"]
Veri = Veri.dropna(axis=0)
Veri
```

3. 2021 ve 2022 veri tablolarını tek bir tablo haline getiriniz.

```python
indirme_listesi = ["XU030.IS","EURTRY=X"]
Veri = pd.DataFrame()

for i in indirme_listesi:
  data = yf.download(i, start="2021-01-01",end="2023-01-01")
  data = data["Adj Close"].rename(i)
  Veri= pd.concat([Veri, data], axis=1)

Veri.columns =["Kapanış (TRY)", "EUR/TRY"]
Veri = Veri.dropna(axis=0)
Veri
```

```python
Veri["Normalize"] = Veri["Kapanış (TRY)"]/ Veri.iloc[0,0]
# Veri["Normalize USD/TRY"] = Veri["USD/TRY"]/ Veri.iloc[0,1]
Veri["Normalize EUR/TRY"] = Veri["EUR/TRY"]/ Veri.loc[Veri.index[0],"EUR/TRY"]
Veri["Degisim"] = Veri.Normalize.diff(1)
Veri["D_yon"] = Veri.Degisim.map(lambda x: "Düşen gün" if x<0 else "Yükselen gün")
Veri["Degisim%"] = (Veri.Degisim/Veri.Normalize)*100
Veri
```

4. Veriyi grafik olarak inceleyiniz:<br>
	• Tabloları px grafiği olarak görselleştiriniz.<br>
	• Tabloları go. grafiği olarak görselleştiriniz.<br>
	• Tabloları subplots grafiği olarak görselleştiriniz.

```python
fig_3 = px.line(Veri, x=Veri.index, y=Veri.Normalize,
        width=1000, height=500,
        title = "EUR/TL 2021 - Normalize (1 ocak 2021)",
        hover_data = {"Normalize": ":.2f", "Kapanış (TRY)": ":.2f",},
        labels = {"Date" : "Tarih", "Normalize" : "BIST 30 Endeksi (TRY)"}
        )

fig_3.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = Veri.index , y= Veri["Normalize"], name="Normalize"))
fig.add_trace(go.Bar(x = Veri.index , y= Veri["Normalize EUR/TRY"], name="Normalize EUR/TRY"))

fig.update_layout(title="Normalize EUR/TRY",
                  xaxis_title = "Tarih", xaxis_dtick="M1",
                  yaxis_title = "Kapanış TRY"

                  )
```

```python
fig = make_subplots(rows=1, cols=2)

fig.add_trace(go.Scatter(x = Veri.index , y= Veri["Normalize"], name="Normalize"),
              row=1, col=1)

fig.add_trace(go.Bar(x = Veri.index , y= Veri["Normalize EUR/TRY"], name="Normalize EUR/TRY"),
              row=1, col=2)

fig.show()
```
